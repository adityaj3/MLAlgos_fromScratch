## Building Vocabulary

```python
%%time
vocab = {}
# Creating a dictionary of words and their frequency
for i in range(len(data)):       # key will be each folder(newsgroups)
    for doc in data[folders[i]]: # value will be each document in that folder
        for word in doc.split(): # For each word in that document
            if word.lower() not in stopWords and len(word.lower()) >= 3:  # not including the stopwords and 3 lettered or smaller words
                if word.lower() not in vocab:
                    vocab[word.lower()] = 1
                else:
                    vocab[word.lower()] += 1
len(vocab)
```

```python
# Sort the dictionary based on frequency of each 'possible' vocabulary word
from operator import itemgetter
sorted_vocab = sorted(vocab.items(), key = itemgetter(1), reverse=True)
```

# Splitting X and Y into Train and Test data

```python
from sklearn import model_selection
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

# Self Implementation of Multinomial Naive Bayes

```python
def fit(X_train, Y_train):
    result = {}
    class_values = set(Y_train)
    for current_class in class_values:
        result[current_class] = {}
        result["total_data"] = len(Y_train)
        current_class_rows = (Y_train == current_class)
        X_train_current = X_train[current_class_rows]
        Y_train_current = Y_train[current_class_rows]
        num_features = X_train.shape[1]
        result[current_class]["total_class_count"] = len(Y_train_current)
        result[current_class]["total_word_count"] = 0
        for j in range(1, num_features + 1):
            result[current_class][j] = (X_train_current[:, j - 1]).sum()
            result[current_class]["total_word_count"] += result[current_class][j]
    return result
```

```python
def probability(dictionary, x, current_class):
    output = np.log(dictionary[current_class]["total_class_count"]) - np.log(dictionary["total_data"])
    num_words = len(dictionary[current_class].keys()) - 2;
    for j in range(1, len(x) + 1):
        xj = x[j - 1]
        count_current_class_with_word_j = dictionary[current_class][j] + 1
        count_current_class = dictionary[current_class]["total_word_count"] + len(x)
        current_xj_probablity = xj*(np.log(count_current_class_with_word_j) - np.log(count_current_class))
        output = output + current_xj_probablity
    return output
```

```python
def predictSinglePoint(dictionary, x):
    classes = dictionary.keys()
    best_p = -1000
    best_class = -1
    first_run = True
    for current_class in classes:
        if (current_class == "total_data"):
            continue
        p_current_class = probability(dictionary, x, current_class)
        if (first_run or p_current_class > best_p):
            best_p = p_current_class
            best_class = current_class
        first_run = False
    return best_class
```

```python
def predict(dictionary, X_test):
    y_pred = []
    for x in X_test:
        x_class = predictSinglePoint(dictionary, x)
        y_pred.append(x_class)
    return y_pred
```

# Inbuilt Multinomial Naive Bayes Implementation

```python
%%time
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix
clf = MultinomialNB()
clf.fit(X_train, Y_train)
Y_pred = clf.predict(X_test)
```
